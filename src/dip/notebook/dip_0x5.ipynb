```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("chips.png")
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
lower = np.array([[40, 50, 50], [20, 100, 100],
                  [170, 120, 70], [90, 100, 100],
                  [2, 100, 100]])
upper = np.array([[80, 255, 255], [40, 255, 255],
                  [180, 255, 255], [130, 255, 255],
                  [8, 255, 255]])
names = ["Verde", "Amarela", "Vermelho", "Azul", "Laranja"]

for i in range(len(lower)):
    mask = cv2.inRange(img_hsv, lower[i], upper[i])

    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, np.ones((8, 8), np.uint8))

    seg_img = cv2.bitwise_and(img, img, mask=mask)
    masked_img = cv2.medianBlur(seg_img, 5)

    cv2_imshow(seg_img)
```
