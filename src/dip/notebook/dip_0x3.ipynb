```python
import cv2
from matplotlib import pyplot as plt
```

#### 1 - Carregar e mostrar cada uma das imagens

```python
imgs = []
path = ["../src/img/strawberries.tif", "../src/img/baboon.png", "../src/img/chips.png",
        "../src/img/em_spectrum2.png", "../src/img/flowers.jpg", "../src/img/hsv_disk.png",
        "../src/img/rgb.png", "../src/img/rgbcube_kBKG.png"]

names = ["strawberries.tif", "baboon.png", "chips.png", "em_spectrum2.png",
         "flowers.jpg", "hsv_disk.png", "rgb.png", "rgbcube_kBKG.png"]
```

```python
[imgs.append(cv2.imread(item)) for item in path]

[cv2.imshow(f'Imagem {i+1}', img) for i, img in enumerate(imgs)]

cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 1.1 Converter BRG para RGB

```python
rgb_imgs = list(map(lambda img: cv2.cvtColor(img, cv2.COLOR_BGR2RGB), imgs))
[cv2.imshow(f'Converted {names[i]}', img) for i, img in enumerate(rgb_imgs)]

cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 2 - Apresentar cada canal de cor (R, G e B) separadamente para cada imagem

```python
def show_channels(name, img, flag=0):

    channels = ["Red", "Green", "Blue"]
    if flag == 0:
        fig, axs = plt.subplots(2, 2, figsize=(10, 10))
        fig.suptitle(name)

        for i in range(4):
            if i == 0:
                axs[0, 0].imshow(img)
                axs[0, 0].set_title("Original Image")
            else:
                axs[i // 2, i % 2].imshow(img[:, :, i - 1], cmap="gray")
                axs[i // 2, i % 2].set_title(channels[i - 1][0])
        return fig
    else:
        #cv2.namedWindow((names[index])
        cv2.imshow(name, img)
        cv2.waitKey(0)
```

```python
[(show_channels(names[index], rgb_imgs[index]), plt.show()) for index in range(len(imgs))]
```

#### 3 - Normalizar cada imagem convertendo para ponto flutuante e dividindo por 255.0.

```python
normalized_imgs = [img.astype("float32") / 255. for img in rgb_imgs]
```

#### 4 - Apresentar as imagens (do tipo float) usando cv2.namedWindow e matplotlib.

```python
[(show_channels(names[index], rgb_imgs[index], 1), plt.show()) for index in range(len(normalized_imgs))]
```
